```python
import lmod
await lmod.load('deepretinotopy/1.0.8')
import os
```

```python
# os.environ["subject"] = "sub-wlsubj001"
os.environ["hemisphere"] = "lh"
os.environ["hemi"] = "L"
os.environ["sub_path"] = '/neurodesktop-storage/nyu_retinotopy_data/freesurfer_test'
```

## Step 1

```python
%%bash
for dir in $sub_path/sub-wlsubj*; do
    echo $dir
    subject=$(basename "$dir")
    wb_shortcuts -freesurfer-resample-prep $dir/surf/"$hemisphere".white $dir/surf/"$hemisphere".pial \
            $dir/surf/"$hemisphere".sphere.reg "$sub_path"/../templates/fs_LR-deformed_to-fsaverage."$hemi".sphere.32k_fs_LR.surf.gii \
            $dir/surf/"$hemisphere".pial.surf.gii $dir/surf/"$subject"."$hemisphere".pial.32k_fs_LR.surf.gii \
            $dir/surf/"$hemisphere".sphere.reg.surf.gii
    echo $subject":successfully convert!"
done
```

## Step 2

```python
%%bash
echo Step2!
for dir in $sub_path/sub-wlsubj*; do
    echo $dir
    subject=$(basename "$dir")
    wb_command -metric-resample "$sub_path"/../nyu_converted/$subject/"$hemisphere".angle.converted.func.gii \
        $dir/surf/"$hemisphere".sphere.reg.surf.gii "$sub_path"/../templates/fs_LR-deformed_to-fsaverage."$hemi".sphere.32k_fs_LR.surf.gii \
        ADAP_BARY_AREA $dir/surf/fs_empirical."$hemisphere".32k_fs_LR.func.gii \
        -area-surfs $dir/surf/"$hemisphere".pial.surf.gii $dir/surf/"$subject"."$hemisphere".pial.32k_fs_LR.surf.gii  
    echo $subject":successfully convert!"
done
```
