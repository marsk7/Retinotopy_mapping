# Convert angle

```python
import os
import lmod
await lmod.load('freesurfer/7.4.1')
import nibabel as nib
import numpy as np

# Convert angles
def convert_angles(infile, outfile):
    template = nib.load(infile)  # This is the new file you generated with mris_convert
    original_angle  = template.agg_data()
    converted_angle = original_angle * 360 / (2*np.pi)

    sum = converted_angle < 180
    minus = converted_angle > 180
    converted_angle[sum] = converted_angle[sum] + 180
    converted_angle[minus] = converted_angle[minus] - 180

    template.agg_data()[:] = converted_angle
    nib.save(template, outfile)
```

```python
# Convert angles of all individuals 
folder_path = '/neurodesktop-storage/nyu_retinotopy_data/nyu_converted/'
hemis = 'lh'

items = os.listdir(folder_path)

for item in items:
    full_path = os.path.join(folder_path, item)
    if os.path.exists(full_path + '/'+hemis+'.'+hemis+'_native_angle.func.gii') and item[:-3] == 'sub-wlsubj' :
        infile = full_path + '/'+hemis+'.'+hemis+'_native_angle.func.gii'
        outfile = full_path + '/'+hemis+'.angle.converted.func.gii'
        
        convert_angles(infile, outfile)
        
        print(item + ': Successfully convert')
```

# Convert angle RH

```python
import os
import lmod
await lmod.load('freesurfer/7.4.1')
import nibabel as nib
import numpy as np

# Convert angles
def convert_angles_rh(infile, outfile):
    template = nib.load(infile)  # This is the new file you generated with mris_convert
    original_angle  = template.agg_data()
    converted_angle = original_angle * 360 / (2*np.pi)

    converted_angle = (converted_angle + 180) % 360
    sum = converted_angle < 180
    minus = converted_angle > 180
    converted_angle[sum] = converted_angle[sum] + 180
    converted_angle[minus] = converted_angle[minus] - 180

    template.agg_data()[:] = converted_angle
    nib.save(template, outfile)
    
# Convert angles of all individuals 
folder_path = '/neurodesktop-storage/nyu_retinotopy_data/nyu_converted/'
hemis = 'rh'

items = os.listdir(folder_path)

for item in items:
    full_path = os.path.join(folder_path, item)
    if os.path.exists(full_path + '/'+hemis+'.'+hemis+'_native_angle.func.gii') and item[:-3] == 'sub-wlsubj' :
        infile = full_path + '/'+hemis+'.'+hemis+'_native_angle.func.gii'
        outfile = full_path + '/'+hemis+'.angle.converted.func.gii'
        
        convert_angles_rh(infile, outfile)
        
        print(item + ': Successfully convert')
```
