# Calculate correlation

```python
import nibabel as nib
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
from python_visualization.visualization import roi
import scipy
import os.path as osp
from python_visualization.rois import ROIs_WangParcels as roi_parcel

def roi_earlyvisualcortex(list_of_labels):
    """Mask for the selection of the region of interest in the surface
    template.

    Args:
        list_of_labels (list): list with the file name (.mat) containing the
            region of interest (from both L and R hemispheres)

    Returns:
        final_mask_L (numpy array): Mask of the region of interest from left
            hemisphere (32492,)

        final_mask_R (numpy array): Mask of the region of interest from
            right hemisphere (32492,)

        index_L_mask (list): Indices of the non-zero elements from
            final_mask_L (number of nonzero elements,)

        index_R_mask (list): Indices of the non-zero elements from
            final_mask_R (number of nonzero elements,)

    """

    # Defining number of nodes
    number_cortical_nodes = int(64984)
    number_hemi_nodes = int(number_cortical_nodes / 2)

    list_primary_visual_areas = np.zeros([len(list_of_labels), 64984])
    for i in range(len(list_of_labels)):
        list_primary_visual_areas[i] = np.reshape(scipy.io.loadmat(
            osp.join(osp.dirname(osp.realpath(__file__)),
                     'python_visualization',
                     list_of_labels[i] + '_V1-3.mat'))[list_of_labels[i]][0:64984],
                                                  (-1))

    final_mask_L = np.sum(list_primary_visual_areas, axis=0)[
                   0:number_hemi_nodes]
    final_mask_R = np.sum(list_primary_visual_areas, axis=0)[
                   number_hemi_nodes:number_cortical_nodes]

    index_L_mask = [i for i, j in enumerate(final_mask_L) if j == 1]
    index_R_mask = [i for i, j in enumerate(final_mask_R) if j == 1]

    return final_mask_L, final_mask_R, index_L_mask, index_R_mask
    
subject = "sub-wlsubj117"
hemisphere = "lh"

# File path
pre_pial_angle = '/neurodesktop-storage/nyu_retinotopy_data/freesurfer_pial2mid/'+subject+ \
            '/deepRetinotopy/'+subject+'.fs_predicted_polarAngle_'+hemisphere+'_curvatureFeat_model.func.gii'
pre_mid_angle = '/neurodesktop-storage/nyu_retinotopy_data/freesurfer_test/'+subject+ \
            '/deepRetinotopy/'+subject+'.fs_predicted_polarAngle_'+hemisphere+'_curvatureFeat_model.func.gii'
pre_white_angle = '/neurodesktop-storage/nyu_retinotopy_data/freesurfer_white2mid/'+subject+ \
            '/deepRetinotopy/'+subject+'.fs_predicted_polarAngle_'+hemisphere+'_curvatureFeat_model.func.gii'

data_path = '/neurodesktop-storage/nyu_retinotopy_data/' 
pial_surf = data_path + 'freesurfer_pial2mid/'+subject+'/surf/'+subject+'.'+hemisphere+'.pial.32k_fs_LR.surf.gii'
mid_surf = data_path + 'freesurfer_test/'+subject+'/surf/'+subject+'.'+hemisphere+'.midthickness.32k_fs_LR.surf.gii'
white_surf = data_path + 'freesurfer_white2mid/'+subject+'/surf/'+subject+'.'+hemisphere+'.white.32k_fs_LR.surf.gii'

# Number of nodes
number_cortical_nodes = int(64984)
number_hemi_nodes = int(number_cortical_nodes / 2)

# visual cortex mask
final_mask_L, final_mask_R, index_L_mask, index_R_mask = roi(['ROI'])
# Early visual cortex
final_mask_L, final_mask_R, index_L_mask, index_R_mask = roi_earlyvisualcortex(['ROI'])
region_of_interest = ['V1', 'V2', 'V3']
areas = ['V1v', 'V2v', 'V3v']
print('Region of interest: ' + str(region_of_interest))

if region_of_interest != 'all' and region_of_interest != 'earlyvisualcortex':
    if region_of_interest == 'V1':
        areas = ['V1v', 'V1d']
    elif region_of_interest == 'V2':
        areas = ['V2v', 'V2d']
    elif region_of_interest == 'V3':
        areas = ['V3v', 'V3d']
    final_mask_L, final_mask_R, index_L_mask, index_R_mask = roi_parcel(areas)
final_mask_L, final_mask_R, index_L_mask, index_R_mask = roi_parcel(areas)
if hemisphere == "lh":
    mask, index = final_mask_L, index_L_mask
elif hemisphere == "rh":
    mask, index = final_mask_R, index_R_mask
    
# Step1 calculate thickness
pial_gii = nib.load(pial_surf)
mid_gii = nib.load(mid_surf)
white_gii = nib.load(white_surf)

pial_coords = pial_gii.agg_data("NIFTI_INTENT_POINTSET")
mid_coords = mid_gii.agg_data("NIFTI_INTENT_POINTSET")
white_coords = white_gii.agg_data("NIFTI_INTENT_POINTSET")

# Applying Mask
pial_coords = pial_coords[mask == 1]
mid_coords = mid_coords[mask == 1]
white_coords = white_coords[mask == 1]

thickness = np.linalg.norm(pial_coords - mid_coords, axis=1)
thickness_w = np.linalg.norm(white_coords - mid_coords, axis=1)

# Step2 Calculate differences 
pial_angle_gii = nib.load(pre_pial_angle)
mid_angle_gii = nib.load(pre_mid_angle)
white_angle_gii = nib.load(pre_white_angle)

pial_polar = pial_angle_gii.darrays[0].data
mid_polar = mid_angle_gii.darrays[0].data 
white_polar = white_angle_gii.darrays[0].data 

# Applying Mask
pial_polar = pial_polar[mask == 1]
mid_polar = mid_polar[mask == 1]
white_polar = white_polar[mask == 1]

angle_diff = np.abs(pial_polar - mid_polar)
# angle_diff[angle_diff <= 100] = 100 # handle as overflow
angle_diff_w = np.abs(white_polar - mid_polar) 

# Visualization
plt.figure(figsize=(6,6))
plt.scatter(thickness, angle_diff, s=1, alpha=0.3)
plt.xlabel("Cortical Thickness")
plt.ylabel("Predicted Polar Angle Difference")
plt.title(f"Correlation between thickness and predicted difference on sub-wlsubj117 based on pial")
plt.show()

# Visualization
plt.figure(figsize=(6,6))
plt.scatter(thickness_w, angle_diff_w, s=1, alpha=0.3)
plt.xlabel("Cortical Thickness")
plt.ylabel("Predicted Polar Angle Difference")
plt.title(f"Correlation between thickness and predicted difference on sub-wlsubj117 based on white")
plt.show()

# Line Chart for Cortical Thickness
plt.figure(figsize=(10, 5))
plt.plot(index_L_mask, thickness, label='Cortical Cortical Thickness', color='blue')
plt.xlabel('Vertex Index')
plt.ylabel('Value')
plt.title('Cortical Thickness')
plt.legend()
plt.show()

# Line Chart for Predicted Polar Angle Difference
plt.figure(figsize=(10, 5))
plt.plot(index_L_mask, angle_diff, label='Predicted Polar Angle Difference', color='red')
plt.xlabel('Vertex Index')
plt.ylabel('Value')
plt.title('Predicted Polar Angle Difference')
plt.legend()
plt.show()
```

```python
import nibabel as nib
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
from python_visualization.visualization import roi
import scipy
import os.path as osp
from python_visualization.rois import ROIs_WangParcels as roi_parcel

def roi_earlyvisualcortex(list_of_labels):
    """Mask for the selection of the region of interest in the surface
    template.

    Args:
        list_of_labels (list): list with the file name (.mat) containing the
            region of interest (from both L and R hemispheres)

    Returns:
        final_mask_L (numpy array): Mask of the region of interest from left
            hemisphere (32492,)

        final_mask_R (numpy array): Mask of the region of interest from
            right hemisphere (32492,)

        index_L_mask (list): Indices of the non-zero elements from
            final_mask_L (number of nonzero elements,)

        index_R_mask (list): Indices of the non-zero elements from
            final_mask_R (number of nonzero elements,)

    """

    # Defining number of nodes
    number_cortical_nodes = int(64984)
    number_hemi_nodes = int(number_cortical_nodes / 2)

    list_primary_visual_areas = np.zeros([len(list_of_labels), 64984])
    for i in range(len(list_of_labels)):
        list_primary_visual_areas[i] = np.reshape(scipy.io.loadmat(
            osp.join(osp.dirname(osp.realpath(__file__)),
                     'python_visualization',
                     list_of_labels[i] + '_V1-3.mat'))[list_of_labels[i]][0:64984],
                                                  (-1))

    final_mask_L = np.sum(list_primary_visual_areas, axis=0)[
                   0:number_hemi_nodes]
    final_mask_R = np.sum(list_primary_visual_areas, axis=0)[
                   number_hemi_nodes:number_cortical_nodes]

    index_L_mask = [i for i, j in enumerate(final_mask_L) if j == 1]
    index_R_mask = [i for i, j in enumerate(final_mask_R) if j == 1]

    return final_mask_L, final_mask_R, index_L_mask, index_R_mask
subject = "sub-wlsubj117"
hemisphere = "lh"

# Curvatue file path
data_path = '/neurodesktop-storage/nyu_retinotopy_data/' 
pial_curv = data_path + 'freesurfer_pial2mid/'+subject+'/surf/'+subject+ \
            '.curvature-midthickness.'+hemisphere+'.32k_fs_LR.func.gii'
mid_curv = data_path + 'freesurfer_test/'+subject+'/surf/'+subject+ \
            '.curvature-midthickness.'+hemisphere+'.32k_fs_LR.func.gii'
white_curv = data_path + 'freesurfer_white2mid/'+subject+'/surf/'+subject+ \
            '.curvature-midthickness.'+hemisphere+'.32k_fs_LR.func.gii'
# Prediction file path
pre_pial_angle = '/neurodesktop-storage/nyu_retinotopy_data/freesurfer_pial2mid/'+subject+ \
            '/deepRetinotopy/'+subject+'.fs_predicted_polarAngle_'+hemisphere+'_curvatureFeat_model.func.gii'
pre_mid_angle = '/neurodesktop-storage/nyu_retinotopy_data/freesurfer_test/'+subject+ \
            '/deepRetinotopy/'+subject+'.fs_predicted_polarAngle_'+hemisphere+'_curvatureFeat_model.func.gii'
pre_white_angle = '/neurodesktop-storage/nyu_retinotopy_data/freesurfer_white2mid/'+subject+ \
            '/deepRetinotopy/'+subject+'.fs_predicted_polarAngle_'+hemisphere+'_curvatureFeat_model.func.gii'

# Number of nodes
number_cortical_nodes = int(64984)
number_hemi_nodes = int(number_cortical_nodes / 2)

# visual cortex mask
final_mask_L, final_mask_R, index_L_mask, index_R_mask = roi(['ROI'])
# Early visual cortex
final_mask_L, final_mask_R, index_L_mask, index_R_mask = roi_earlyvisualcortex(['ROI'])
region_of_interest = ['V1', 'V2', 'V3']
areas = ['V2v']
# print('Region of interest: ' + str(region_of_interest))
final_mask_L, final_mask_R, index_L_mask, index_R_mask = roi_parcel(areas)
if hemisphere == "lh":
    mask, index = final_mask_L, index_L_mask
elif hemisphere == "rh":
    mask, index = final_mask_R, index_R_mask
    
# Step1 calculate thickness
pial_curv_data = nib.load(pial_curv).darrays[0].data
mid_curv_data = nib.load(mid_curv).darrays[0].data
white_curv_data = nib.load(white_curv).darrays[0].data

# Applying Mask
# pial_curv_data = pial_curv_data[final_mask_R == 1]
# mid_curv_data = mid_curv_data[final_mask_R == 1]
# white_curv_data = white_curv_data[final_mask_R == 1]
pial_curv_data = pial_curv_data[mask == 1]
mid_curv_data = mid_curv_data[mask == 1]
white_curv_data = white_curv_data[mask == 1]

pial_curv_diff = np.abs(pial_curv_data - mid_curv_data)
white_curv_diff = np.abs(white_curv_data - mid_curv_data)

# Step2 Calculate differences 
pial_angle_gii = nib.load(pre_pial_angle)
mid_angle_gii = nib.load(pre_mid_angle)
white_angle_gii = nib.load(pre_white_angle)

pial_polar = pial_angle_gii.darrays[0].data
mid_polar = mid_angle_gii.darrays[0].data 
white_polar = white_angle_gii.darrays[0].data 

# Applying Mask
# pial_polar = pial_polar[final_mask_R == 1]
# mid_polar = mid_polar[final_mask_R == 1]
# white_polar = white_polar[final_mask_R == 1]
pial_polar = pial_polar[mask == 1]
mid_polar = mid_polar[mask == 1]
white_polar = white_polar[mask == 1]

angle_diff = np.abs(pial_polar - mid_polar)
# angle_diff[angle_diff <= 100] = 100 # handle as overflow
angle_diff_w = np.abs(white_polar - mid_polar) 

# Visualization
plt.figure(figsize=(6,6))
plt.xlim(-2, 2)
plt.scatter(pial_curv_diff, angle_diff, s=1, alpha=0.3)
plt.xlabel("Pial curvature differences")
plt.ylabel("Predicted Polar Angle Difference")
plt.title(f"Correlation between thickness and predicted difference on {subject} based on pial")
plt.show()

# Visualization
plt.figure(figsize=(6,6))
plt.scatter(white_curv_diff, angle_diff_w, s=1, alpha=0.3)
plt.xlabel("White curvature differences")
plt.ylabel("Predicted Polar Angle Difference")
plt.title(f"Correlation between thickness and predicted difference on {subject} based on white")
plt.show()

# Line Chart for Cortical Thickness
plt.figure(figsize=(10, 5))
plt.plot(index, pial_curv_diff, label='Pial curvature differences', color='blue')
plt.xlabel('Vertex Index')
plt.ylabel('Pial curvature differences')
plt.title('Correlation Between Pial Curvature Differences and Vertices')
plt.legend()
plt.show()

# Line Chart for Predicted Polar Angle Difference
plt.figure(figsize=(10, 5))
plt.plot(index, angle_diff, label='Predicted Polar Angle Difference', color='red')
plt.xlabel('Vertex Index')
plt.ylabel('Predicted Polar Angle Difference')
plt.title('Correlation Between Predicted Polar Angle Differences and Vertices')
plt.legend()
plt.show()
```

## Calculate thickness

```python
import nibabel as nib
import numpy as np
from nilearn import plotting

file_path = '/neurodesktop-storage/nyu_retinotopy_data/freesurfer_test/sub-wlsubj117/surf/'
# pial_path = file_path + 'lh.pial.surf.gii'
pial_path = file_path + 'sub-wlsubj117.lh.pial.32k_fs_LR.surf.gii'
white_path = file_path + 'sub-wlsubj117.lh.white.32k_fs_LR.surf.gii'
# midthick_path = file_path + 'lh.midthickness.surf.gii'
midthick_path = file_path + 'sub-wlsubj117.lh.midthickness.32k_fs_LR.surf.gii'

# lh_sp_path = file_path + 'lh.sphere'
lh_sp_path = '/neurodesktop-storage/nyu_retinotopy_data/templates/fs_LR-deformed_to-fsaverage.L.sphere.32k_fs_LR.surf.gii'

pial = nib.load(pial_path)
white = nib.load(white_path)
midthick = nib.load(midthick_path)

# load the coordinate data
pial_coords = pial.agg_data('NIFTI_INTENT_POINTSET')
white_coords = white.agg_data('NIFTI_INTENT_POINTSET')
midthick_coords = midthick.agg_data('NIFTI_INTENT_POINTSET')
pial_faces = pial.agg_data('NIFTI_INTENT_TRIANGLE')

# calculate thickness
cortical_thickness = np.linalg.norm(pial_coords - midthick_coords, axis=1)
print(cortical_thickness.shape)

# Visualization
view = plotting.view_surf(
    surf_mesh=lh_sp_path,
    # surf_mesh=midthick_path,
    surf_map=cortical_thickness,
    cmap='rainbow'
)
view
```
